```python
import matplotlib.pyplot as plt
import numpy as np
import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
model = load_model('model/image_cnn_model.h5')
```

```python
class_names = ['glass', 'metal', 'paper', 'plastic', 'trash']
```

```python
test_path = "D:/Recycling_Keras/sample/sample27.jpg"
img = keras.preprocessing.image.load_img(
    test_path, target_size=(300, 300)
)
img_array = keras.preprocessing.image.img_to_array(img)
img_array = tf.expand_dims(img_array, 0) # Create a batch

predictions = model.predict(img_array)
score = tf.nn.softmax(predictions[0])

image = plt.imread(test_path)
plt.imshow(image)
plt.show()

print(
    "This image most likely belongs to {} with a {:.2f} percent confidence."
    .format(class_names[np.argmax(score)], 100 * np.max(score))
)
```

```python
f = open("D:/Recycling_Keras/result.txt", 'w')
f.write(class_names[np.argmax(score)])
f.close()
```
